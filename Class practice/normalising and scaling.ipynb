```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('Data/Churn_Modelling.csv')
data.head()
```

```python
# Normalizing variables https://beginanalyticsblog.wordpress.com/2019/07/10/normalizing-numeric-variables-in-python/

#Normalize variables between 0 & 1

VariableList = ['Age','Tenure','Balance','NumOfProducts','EstimatedSalary']
normVariableName = ['Age_norm','Tenure_norm','Balance_norm','NumOfProducts_norm','EstimatedSalary+norm']

scaler = MinMaxScaler()

for i in range(0,len(data)):
    for j in range(0,len(VariableList)):
        temp = data[VariableList[j]]
        temp1 = temp.values.reshape(-1,1)
        scaler.fit(temp1)
        norm =scaler.transform(temp1)
        data.set_value(i,normVariableName[j],norm[i])
data
# The values of all the variables now ranges between 0 & 1.
```

```python
# Binning https://pbpython.com/pandas-qcut-cut.html

sns.set_style('whitegrid')


df = data.groupby(['CustomerID', 'Surname'])['EstimatedSalary'].sum().reset_index()

df['EstimatedSalary'].plot(kind='hist')

df['EstimatedSalary'].describe()

pd.qcut(df['EstimatedSalary'], q=4)

df['quantile_ex_1'] = pd.qcut(df['EstimatedSalary'], q=4)
df['quantile_ex_2'] = pd.qcut(df['EstimatedSalary'], q=10, precision=0)

df['quantile_ex_1'].value_counts()
df['quantile_ex_2'].value_counts()
```
