```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.linspace(-1,1,5)
x = np.arange(5)

fig = plt.figure()
ax = fig.subplots() # the ax variable acts as a container which holds our plots
line1 = ax.plot(x,y)

y2 = np.linspace(-2,2,5)
line2 = ax.plot(x,y2)
```

```python

# if you want subplots: you can do it like this:

number_rows = 3
number_columns = 2
fig2 = plt.figure()
ax2 = fig2.subplots(number_rows,number_columns) # so ax2 is jut a numpy array holding matplotlib.axes bojects
ax2[0,1].plot(x,y) # this puts a plot in row 1 and column 2
```

```python
# Every aspect of your figure can change, this is why it is important to store every graphical object to a separate variable, one for the figure, one for the axes, one for each line etc then you can manipulate programmatically which figure to close or which axes to reset eg:

plt.close(fig) # closes the figure corresponding to variable "fig"
plt.close("all") # closes all figures
plt.gcf() # get current figure, provides a reference to current selected figure
plt.gca() # get current figure, provides a reference to the current selected axes
plt.clf() # clear current selected figure
plt.cla() # clear current selected axes

# Current is the last selected, NOT the last generated figure or axes
```

```python
y1 = np.arange(0,110,10)
y2 = np.random.random(11)
x = np.arange(11)
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,y1) # line plot
ax[0,1].scatter(x,y2) # scatter plot
ax[1,0].bar(x,y1) # horizontal
ax[1,1].barh(x,y1) # horizontal bars
```

```python
x = np.arange(0,4*np.pi,0.05)
ycos = np.cos(x)
ysin = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x,ycos, 'b-', label="cos(x)")
ax.plot(x,ysin, 'r--', label ="sin(x)")

ax.set_title("Trigonometric functions") #sets a title
ax.set_xlabel("0 to 4pi") # set the x axis title
ax.set_ylabel("cos(x) and sin(x)") # sets the y axis title
ax.legend() # adds the legen
```

```python
# Adding horizontal or vertical lines to a plot
fig, ax = plt.subplots(1,2)

ax[0].axhline(0.4,linestyle='--', color='blue')

# Set aces limit:

ax[0].set_ylim([0.1,0.9])
ax[1].set_xlim([0.1,0.9])

# Set title to figure and subplots
fig.suptitle("Horizontal and vertical lines", fontsize=14) # overall title
ax[0].set_title('Horizontal lines', fontstyle='italic') # title of first plot
ax[1].set_title('Vertical ines', color='green',fontname='Arial') # title of second plot
```

```python
x = np.array([2,4,6,7,4,5,3,7,2,4,5,6,3])
y = np.array([3,3,4,3,2,4,5,2,3,5,2,6,7])

fig, ax = plt.subplots(1,3)
ax[0].scatter(x,y)
ax[0].set_title('default')

ax[1].scatter(x,y,50,marker='+')
ax[1].set_title('size = 50, style = +')

crosses = ax[2].scatter(x,y,200,marker='+',linewidth=3)
bullets = ax[2].scatter(x,y,50, marker='o', color='black')
bullets.set_edgecolors('red')
bullets.set_linewidth(1.5)
ax[2].set_title('mixed')
```

```python
# 3d scatter plot

z = np.array([0,1,3,4,1,2,3,5,3,5,7,3,5])
fig = plt.figure()
ax = np.array([fig.add_subplot(1,3,1,projection='3d'),
               fig.add_subplot(1,3,2,projection='3d'),
               fig.add_subplot(1,3,3,projection='3d')])

ax[0].scatter(x,y,z)
ax[0].set_title('default')

ax[1].scatter(x,y,z, s=50, marker='+')
ax[1].set_title('size = 50,style = +')

crosses = ax[2].scatter(x,y,z,s=200, marker='+',linewidth=3)
bullets = ax[2].scatter(x,y,z, s=50, marker='o', color='black')
bullets.set_edgecolors('red')
bullets.set_linewidth(1.5)
ax[2].set_title('mixed')
```

```python
# Bar plot
people = ["Student A", "Student B"]
studentA = np.array([90,50,80,40])
studentB = np.array([75,45,60,95])
x = np.arange(len(studentA))

fig, ax = plt.subplots(1,2)

ax[0].bar(x,studentA,width=0.3)
ax[1].bar(x,studentB,width=0.3)

for i in range(len(ax)):
    ax[i].set_ylim([0,100])
    ax[i].set_title(people[i])
    ax[i].set_xlabel("exercises")
    ax[i].set_ylabel("mark (%)")
    ax[i].set_xticks([0,1,2,3])
    ax[i].set_xticklabels(['ex1','ex2','ex3','ex4'])
```

```python
# grouped bar plot
fig, ax = plt.subplots()
width = 0.3
s1bars = ax.bar(x-width/2,studentA,width, label = "Student A")
s2bars = ax.bar(x+ width/2,studentB, width, label = "Student B")


ax.set_ylim([0,100])
ax.set_title("Marks for student A and student B")
ax.set_xlabel("exercises")
ax.set_ylabel("mark (%)")
ax.set_xticks([0,1,2,3])
ax.set_xticklabels(['ex1','ex2','ex3','ex4'])

ax.legend()

for i in range(0,4):
    s1bars[i].set_linewidth(3.5)
    s2bars[i].set_linewidth(3.5)

for i in range(0,4):
    if studentA[i]<50:
        s1bars[i].set_edgecolor('red')
    if studentB[i]<50:
        s2bars[i].set_edgecolor('red')
        
   
```

```python
# Adding shapes to a danrom grid:

fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_xlim([-10,10])
ax.axis('scaled')

# the first rectangle:
start_x = -2.5
start_y = 2
width = 5
height = 6
facecolor = 'gray'

rectangle= plt.Rectangle((start_x,start_y), width, height, fc=facecolor)
ax.add_patch(rectangle)

# the second rectangle
rectangle = plt.Rectangle((-0.5,2), 1,2,fc="brown")
ax.add_patch(rectangle)

# some yellow circles
start_x = -1
start_y = 6
radius = 0.5
facecolor = 'yellow'
edgecolor = 'black'
linewidth = 1

circle = plt.Circle((start_x,start_y), radius,
fc=facecolor, ec=edgecolor, lw=linewidth)
ax.add_patch(circle)

circle = plt.Circle((1,6), 0.5, fc="yellow", ec="black", lw=1)
ax.add_patch(circle)

# but for trianles need a new submodule:
from matplotlib.patches import Polygon
start_xy = np.array([-2.5,8])
end_xy = np.array([2.5,8])
top_xy = np.array([0,10])

triangle = Polygon(np.vstack([start_xy, end_xy, top_xy]), color="red")
ax.add_patch(triangle)
```

```python
# Pandas
import pandas as pd
```

```python

# Creating
df = pd.DataFrame(
        {"a": [1,2,3],
         "b": [4,5,6],
         "c": [7,8,9]},
        index = [1,2,3])

df = pd.DataFrame(
        [[1,4,7],
         [2,5,8],
         [3,6,9]],
        index = [1,2,3],
        columns = ['a','b','c'])

dfMultiIndex = pd.DataFrame(
        {"a": [1,2,3],
         "b": [4,5,6],
         "c": [7,8,9]},
        index = pd.MultiIndex.from_tuples(
                [('d',1),('d',2),('e',2)],
                names = ['x1','x2']))


# indexing can be done with loc (their names) or iloc (using the numbers)

#These two both take the same item
print(df.loc[1,'c'])
print(df.iloc[0,2])

# these also take the same row from column 2
print(df.loc[2,['b','c']])
print(df.iloc[1,1:3])

# this takes the first column
print(df.loc[:,'a'])
print(df.iloc[:,0])
```

```python
# concat is used to append either rows or columns

# this adds extra the columns
pd.concat([df,df], axis=1)

# this adds extra rows
pd.concat([df,df])
```

```python
# you can sort using sort_values

# This sorts on column a, the first
df.sort_values('a',ascending=False)

# this sorts the roww order
df.sort_values(1,axis=1,ascending=False)
```

```python
# you can use drop to delete rows or columns

# deletes column c
df.drop(columns='c')

# drops rows 1 and 3
df.drop([1,3])
```

```python
import seaborn as sns
```

```python
"""
---------------
Working Example
---------------
"""
tipsData = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')

tipsData.head()
```

```python
# checking how much is empty
tipsData.info()
tipsData.isnull().sum()
```

```python
tipsData.describe()
```

```python
## show the total bill, tip, size for each day

tipsData.groupby(['day']).sum()
# be nice to stick a pie chart on this data
```

```python
# Checking for correlation between variables
tipsData.corr()

# could use seaborn heatmap to display this
sns.heatmap(tipsData.corr(),annot=True)
plt.figure(figsize=(10,5))
plt.show()
```

```python
# how many people went to the bar by sex by day
tipsData.groupby(['sex','day']).sum()
sns.countplot(x = 'sex', data = tipsData, hue = 'day')
plt.title ("Count of number of boys and girls by day")
```

```python
# Group by day and count rows per day
tipsData.groupby(['day']).count()

# Percentage of tips per day 
tipDayPercentage = 100 * tipsData.groupby(['day']).sum()['tip'] / tipsData.groupby(['day']).sum()['total_bill']

# Convert to dataframe
tipDayPercentage = tipDayPercentage.to_frame('tip(%)').reset_index()

# Pie plot
days = pd.DataFrame(tipsData['day'].value_counts())
days.reset_index(inplace=True)
fig, ax = plt.subplots()
days.plot(kind='pie',y='day',labels=days['index'],autopct='%1.2f', ax=ax)
```

```python
# how many people went to the bar by sex by day
print(tipsData.groupby(['smoker','day']).sum())
sns.countplot(x = 'smoker', data = tipsData, hue = 'day')
plt.title ("Count of smokers vs non smokers by day")
```

```python
#box plot
sns.boxplot(x= 'day', y = 'tip', data = tipsData)
```
