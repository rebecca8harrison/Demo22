```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Data/car_data.csv')

# Description of the dataset:
# columns: name,year,selling_price,km_driven,mileage,engine,max_power
# column 0: name of the car, string
# column 1: year car was built, integer
# column 2: selling price, integer
# Column 3: km_driven integer
# Column 4: mileage float integer
# Column 5: engine size
# Column 6: max power
```

```python
# select the outcome variable Y
Y = data.iloc[:, 2].values.reshape(-1, 1)
# Create an empty plot to display all the relationships
fig, ax = plt.subplots(2,2)

# Run linear regression for Max power
X_maxpower = data.iloc[:, 6].values.reshape(-1, 1)
linear_regressor_1 = LinearRegression()  # create a linear regression object
linear_regressor_1.fit(X_maxpower, Y)  # perform the linear regression
Y_pred_maxpower = linear_regressor_1.predict(X_maxpower)  # make the predictions based on the model

# Run linear regression for Engine size
X_enginesize = data.iloc[:, 5].values.reshape(-1, 1)
linear_regressor_2 = LinearRegression()  # create a linear regression object
linear_regressor_2.fit(X_enginesize, Y)  # perform the linear regression
Y_pred_enginesize = linear_regressor_2.predict(X_enginesize)  # make the predictions based on the model

# Run linear regression for miles
X_miles = data.iloc[:, 4].values.reshape(-1, 1)
linear_regressor_3 = LinearRegression()  # create a linear regression object
linear_regressor_3.fit(X_miles, Y)  # perform the linear regression
Y_pred_miles = linear_regressor_3.predict(X_miles)  # make the predictions based on the model

# Run linear regression for km driven
X_km = data.iloc[:, 3].values.reshape(-1, 1)
linear_regressor_4 = LinearRegression()  # create a linear regression object
linear_regressor_4.fit(X_km, Y)  # perform the linear regression
Y_pred_km = linear_regressor_4.predict(X_km)  # make the predictions based on the model

# Run linear regression for Year
X_year = data.iloc[:, 1].values.reshape(-1, 1)
linear_regressor_5 = LinearRegression()  # create a linear regression object
linear_regressor_5.fit(X_year, Y)  # perform the linear regression
Y_pred_year = linear_regressor_5.predict(X_year)  # make the predictions based on the model



# Plot for max power
ax[0,0].plot(X_maxpower,Y_pred_maxpower, color='green') # line plot
ax[0,0].scatter(X_maxpower,Y) # scatter plot
ax[0,0].set_title('Max power')
ax[0,0].set_ylabel('Price')

# Plot for Engine size
ax[0,1].plot(X_enginesize,Y_pred_enginesize, color='green') # line plot
ax[0,1].scatter(X_enginesize,Y) # scatter plot
ax[0,1].set_title('Engine size')
ax[0,1].set_ylabel('Price')

# Plot for Mileage
ax[1,1].plot(X_miles,Y_pred_miles, color='green') # line plot
ax[1,1].scatter(X_miles,Y) # scatter plot
ax[1,1].set_title('Mileage')
ax[1,1].set_ylabel('Price')

# Plot for Km driven
#ax[1,0].plot(X_km,Y_pred_km, color='green') # line plot
#ax[1,0].scatter(X_km,Y) # scatter plot
#ax[1,0].set_title('Km drive')
#ax[1,0].set_ylabel('Price')

# Plot for Year car built
ax[1,0].plot(X_year,Y_pred_year, color='green') # line plot
ax[1,0].scatter(X_year,Y) # scatter plot
ax[1,0].set_title('Year')
ax[1,0].set_ylabel('Price')


fig.suptitle("Relationship between variables and price", fontsize=14) # overall title


```

```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv('Data/food_data.csv')  # load our data
# The columns are: Food,Calories,Calories from Fat,Total Fat,Total Fat,Sodium,Sodium,Potassium,Potassium,Total Carbo-hydrate,Total Carbo-hydrate,Protein,Vitamin A,Vitamin C,Calcium,Iron,Food Type
data.index = data.iloc[:,0] # Name rows as food for reference
data_to_use = data.iloc[:,1:16] # these are the independent variables or X
data_to_target = data.iloc[:,16] # this is the dependent / outcome variable or Y
```

```python
data.corr()
# looks like there are loads that are highly correlated to remove
```

Need to check for correlation before running - as highly ocrrelated features are counted twice in the model so their importance is overestimated

Naive bayes has no variance to minimize therefore ensemble methods will not offer anything
you can add prior probabilities the the training points are fewer than the features
need to check all the features are in a normal distribution, or log transform etc to make sure they are

```python
X_train, X_test, y_train, y_test = train_test_split(data_to_use, data_to_target, test_size=0.3, random_state=21) # 70% training, 30% test
```

```python


gaussian_model = GaussianNB() # Gaussian model

gaussian_model.fit(X_train, y_train) # Training the model

prediction = gaussian_model.predict(X_test) # Predict the response

result = {'food_type':y_test, 'pred':prediction} 
result_df = pd.DataFrame(result) # creates a new df

print("Accuracy:", metrics.accuracy_score(y_test, prediction)) # Checking NB accuracy
```

Below is a logistic regression in python

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('Data/candy-data.csv') # load our data

X = data.iloc[:, [3,11]].values # using these variables to determine if its a chocolate or not
y = data.iloc[:, 1].values # this is to predict if the candy is a chocolate

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train) # Scaling  training set
X_test = scaler_X.transform(X_test) # Scaling test set

logisticRegression_model = LogisticRegression(random_state=0) # Logistic Regression model
logisticRegression_model.fit(X_train, y_train) # Training the model

prediction = logisticRegression_model.predict(X_test) # Predict the response

confusion_matrix = confusion_matrix(y_test, prediction) # Create a confusion matrix this is like a fals poseitive, false neg matrix
print("Accuracy:", (confusion_matrix[0][0] + confusion_matrix[1][1]) / 22) # Checking LR accuracy
```

Advantages of logistic regression:

fast and easy to implement and interpret

works ok even with correlated features

No assumptons about distributions of classes in feature space

model coeeficients can be viewed as indicators of feature importance

Disadvantages:

complete seperation: if we have an amazing feature, that perfectly separates the the outcome then it doesn't work

if number of observations < number of features, then may have over fitting. so needs lots of observations

assumptions of linearity between the dependent variable and the independent variables therefore no linear problems can't be solved with log regression, so it is not well suited to model complex relationships
